### Es 2 
Genera un arrai di numpy contenente i primi 100 numeri primi.

```python
def isPrime(num):
    if num>1:
        return False
    for i in range (2, math.isqrt(num)+1):
        if num%i==0:
            return False
    return True
```

### Es 7
Disegna un bersaglio 10x10: le cornici concetriche hanno valori crescenti verso il centro (quella più esterna ha valori pari a 1)

```python
size=10
matrix_target=np.zeros((size,size))
#print(matrix_target)
for i in range(size//2):
    matrix_target[i:len(matrix_target)-i, i:len(matrix_target)-i]+=1
#print(matrix_target)
plt.imshow(matrix_target)
```

Simula diecimila partite ciascuna da 1000 lanci di una moneta in quante si riscontra lmento il 55% di croci sui 1000 lanci?



```python
n_throw=1000
n_games=10000
percent=0.55
count=0
for _ in range(n_games):
    game=np.random.choice(["Testa", "Croce"], n_throw)
    if(np.count_nonzero(game=="Croce")>n_throw*percent):
        count+=1
print(count)
```

```python
# alra versione 
games=np.random.choice(["Testa", "Croce"], (n_games, n_throw))
tail=games=="Croce"
```

```python
np.count_nonzero(tail) # quante volte appare croce in tutta la matrice
```

```python
np.count_nonzero((np.count_nonzero(tail, axis=1)/n_throw)>percent) # quante volte abbiamo ottenuto true in ognuno dei miei vattori da 10000
```
