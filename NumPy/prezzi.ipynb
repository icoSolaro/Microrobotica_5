## Simuliamo il prezzo giornaliero di una azione
Questa approssiamazione di prezzo Luoi Bachelier. primo matematico che approssima il modello del mercato

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_GIORNI= 100
variazioni_prezzo=np.random.normal(2, 7, N_GIORNI) # se aumento media aumento l'incremento, se aumento deviazione standard aumenta volatilià
prezzo_iniziale = 50
prezzo_giornaliero=prezzo_iniziale+np.cumsum(variazioni_prezzo) # è un array che contiene il prezzo per giorno calcolando in base al prezzo iniziale più la somma di tutte le variazioni fino a quel giorno
```

```python
fig, ax= plt.subplots(figsize=(5,2))
ax.plot(prezzo_giornaliero)
ax.set_xlabel("Data")
ax.set_ylabel("Prezzo")
```

```python
#trovare massimi e minimi di prezzo giornaliero:
giornaliero_prev=np.roll(prezzo_giornaliero, -1)
giornaliero_next=np.roll(prezzo_giornaliero, 1)

giornalieri_without_estremi = prezzo_giornaliero[1:-1]
prev_without_estremi = giornaliero_prev[1:-1]
next_without_estremi = giornaliero_next[1:-1]

max=giornalieriWithoutEstremi[(
    (giornalieri_without_estremi > prev_without_estremi) & 
    (giornalieri_without_estremi > next_without_estremi)
)]
print(max)
```
