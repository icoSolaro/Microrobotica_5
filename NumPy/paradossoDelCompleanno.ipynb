#### fare una funzione che dato il numero alunni e nclassi con default 1000 che calcoli la probabilità di 2 alunni di avere stesso compleanno

```python
def prob_duplicate(n_alunni, n_classi=1000):
    classi=np.random.randint(1, 366, (n_classi, n_alunni))
    count = 0
    for i in range(n_classi):
        if len(classi[i])!=len(np.unique(classi[i])):
            count=count+1
    return count/n_classi
```

### grafici
con un array di numero di alunni che aumenta progressivamente e valutiamo l'aumento della percentuale

```python
array_n_alunni=np.arange(12, 28)
array_n_alunni
```

```python
prob=np.array([prob_duplicate(i) for i in array_n_alunni])
```

```python
fig, ax=plt.subplots(figsize=(7, 4))
ax.plot(array_n_alunni, prob)
ax.set_xlabel("Numero di alunnu")
ax.set_title(r"Probabilità che almeno 2 alunni siano nati lo stesso giorno")
ax.hlines(0.5, 12, 27, colors="red", linestyles="dotted")
ax.grid()
plt.show()
```
