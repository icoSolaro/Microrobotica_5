## Calcolare il $\pi$ con il metodo Monte Carlo
Nasce a los alamos con von newman, fermi... metodi per generazione di numeri pseudocasuali. Servono a calcolare integrale con tante dimensioni: integrale dx, dy, dx, du... Tante varialibi poco controllabili. In finanza stimare andamento con tante variabili. 
Non lo usiamo per far convergere lentamente il $\pi$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_LANCI=15_900_000
RAGGIO = 1. #m
x=np.random.uniform(-RAGGIO, RAGGIO, N_LANCI)
y=np.random.uniform(-RAGGIO, RAGGIO, N_LANCI)

inside_circle=x**2+y**2<RAGGIO**2
n_inside_circle=np.count_nonzero(inside_circle)

print(f"il valore stimato è: {4*n_inside_circle/N_LANCI}")
```

```python
fig, ax= plt.subplots(figsize=(5, 5))

pi_stimato=4*n_inside_circle/N_LANCI

ax.scatter(x, y, c=inside_circle, s=1., cmap="bwr")
ax.set_xlabel("$x$")
ax.set_xlabel("$y$")
ax.set_title(r"il valore di $\pi$ stimato con %s è: %s"%(N_LANCI, pi_stimato))

plt.show()
```

### Esercizio funzione errore pi e n° lanci
Creare una funzione che stimi il $\pi$ dato il numero di lanci. tracciare un gradico dell'errore nella stima di $\pi$ in funzione dei lanci

```python
def stimaPi_perNlanci(nLanci):
    x=np.random.uniform(-RAGGIO, RAGGIO, nLanci)
    y=np.random.uniform(-RAGGIO, RAGGIO, nLanci)

    inside_circle=x**2+y**2<RAGGIO**2
    n_inside_circle=np.count_nonzero(inside_circle)
    return 4*n_inside_circle/nLanci
```

```python
x=np.linspace(1, 900_000, 900, dtype=int)
stimaPerLancio = np.array([stimaPi_perNlanci(n) for n in x])
stimaPerLancio
```

```python
y = np.abs(stimaPerLancio - np.pi)
#y
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y)
ax.set_xlabel("Numero di lanci")
ax.set_ylabel(r"Errore assoluto $|\pi_{stima} - \pi|$")
ax.set_title(r"Errore nella stima di $\pi$ in funzione del numero di lanci")
ax.grid(True, alpha=0.3)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.loglog(x, y)  # Scala logaritmica su entrambi gli assi
ax.set_xlabel("Numero di lanci")
ax.set_ylabel(r"Errore assoluto $|\pi_{stima} - \pi|$")
ax.set_title(r"Errore nella stima di $\pi$ in funzione del numero di lanci (scala log-log)")
ax.grid(True, alpha=0.3, which='both')
plt.show()
```

Esempio di calcolo $\pi$ cerchio inscritto in un quadrato con origine negli assi. abbiamo il raggio r e il lato risulta 2r. il numero di freccette che cadono nel cerchio / numero di lanci quindi numero di quelle che cadono nel quadrato. Il numero di freccette lanciate è direttamente proporzionale all'area del quadrato e del cerchio. La relazione tra le 2 aree è uguale alla relazione tra i due numeri di freccette nell'area. Se facessi infiniti lanci beccherei tutti gli infiniti punti dell'area. Il rapporto tra le 2 aree di infiniti punti avrei un approssimazione perfetta del $\pi$

$\lim_{N_l->\infty} \frac{Nc}{Nl}=\frac{\pi}{4}$
