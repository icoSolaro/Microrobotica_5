## Stima l'integrale definito di una funzione con numpy e il metodo MonteCarlo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x + 2*np.sin(2*x) + 0.5*np.cos(5*x)
```

```python
x_fun= np.linspace(-6, 15, 100000)
y_fun=f(x_fun)
```

### integrale definito nell'intervallo [4, 6]
La funzione nell'intervallo raggiunge un valore massimo di circa 7.82 utilizzo 8 come lato del rettangolo

```python
N_PUNTI=1_000
ESTREMO1=4
ESTREMO2=6

x=np.random.uniform(ESTREMO1, ESTREMO2, N_PUNTI)
y=np.random.uniform(0, 8, N_PUNTI)

sottesa =  f(x) > y 
n_punti_sottesi=np.count_nonzero(sottesa)
integrale=(n_punti_sottesi / N_PUNTI) * (ESTREMO2 - ESTREMO1) * 8

print(f"il valore dell'integrale definito [{ESTREMO1}, {ESTREMO2}] è: {integrale}")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
fig.set_facecolor("pink")
ax.plot(x_fun, y_fun, label="$f(x)$", linestyle="-.")
ax.legend()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_title("y=f(x)")
ax.hlines(0, -6, 15, colors="black")
ax.hlines(8, -6, 15, colors="red", linestyles="dotted")
ax.vlines(4, 20, -5, colors="black")
ax.vlines(6, 20, -5, colors="black")
ax.grid(True)
ax.scatter(x, y, c=sottesa, s=0.3, label="punto Monte Carlo" )
plt.show()
```
