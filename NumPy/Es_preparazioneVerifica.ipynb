### Esercizio 1
In una classe di 27 alunni ogni alunno riceve 10 voti di una materia assegnate a caso. Le valutazioni possibili sono comprese tra 2 e 10, inclusi gli estremi, e non sono ammessi mezzi voti. Utilizza np per simulare i 27 x 10 voti. Calcola quanti alunni hanno un voto medio > 6

```python
import numpy as np
```

```python
N_ALUNNI=27
N_VALUTAZIONI=10
V_MIN=2
V_MAX=10
matrix_voti=np.random.randint(V_MIN, V_MAX+1, (27, 10))
matrix_voti
```

```python
medie=np.mean(matrix_voti, axis=1)
medie
```

```python
medieSufficienti=np.count_nonzero(medie>6)
medieSufficienti
```

```python
indici=np.arange(1, N_ALUNNI+1)
indici
```

```python
#vogliamo il numero degli alunni che hanno media sufficiente
indici[medie>6]
```

### Esercizio 2
Una np per creare l'immagine seguente compasta da n righe e 2n colonne. L'intesità del colore è casuale nella metà sinistra della matrice, a scacchiera nella metà destra. Non puoi suare if, cilco e liste 

```python
import matplotlib.pyplot as plt
```

```python
N=10

colors=np.random.random((N,N))

matrix=np.zeros((N, 2*N))

matrix[::2, N::2]=1.
matrix[1::2, N+1::2]=1.
matrix[:, :N]=colors

plt.imshow(matrix)
```
