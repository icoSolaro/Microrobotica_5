### array multidimensionali

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# axis 0=> colonne
matrix=np.arange(10)   #crea sequenze di numeri interi 
matrix
```

```python
matrix.shape
```

```python
matrix.dtype()
```

```python
 # axis 0 => righe 
# axis 1=> colonne

matrix=np.arange(10).reshape((2, 5))    #vado a riscrivere i miei elementi in righe e colonne
matrix
matrix.shape
```

```python
# axis 0 => righe
#axis 1=>colonna
#axis 2=>profondità

matrix=np.arange(10).reshape((2, -1))  #con il -1 calcola in automatico il valore che manca  
matrix
```

## Indicizzazione degli array
### Indicizzazione Pyhton


```python
array1d=np.arange(8)
```

```python
array1d[4] #indicizzazione python
```

```python
array1d[2:7:2]   #slicing py con passo
```

```python
#prima novità
array1d[2:7:2]=99 #broadcast: su ogni elemento dell'array di cui vado a indicizzare con slicing assegna il valore 99
array1d
```

## Indicizzazione booleana

```python
# gli array di booleani possono essere usati come indici
array1d
```

```python
# al posto dei 99 voglio scrivere 1000:
bool_index=array1d==99
bool_index
```

```python
array1d[bool_index]=1000    #la shape dell'array che voglio indicizzare e dell'array booleano devono essere uguali
array1d
```

`array1d[bool_index]` in NumPy si chiama **view** perchè non sta allocando un nuovo array, ma semplicemente va a visualizzare ciò che l'array bool ha true

```python
matrix_ones=np.ones((8,8))
matrix_ones
```

```python
matrix_ones[::2]=0
matrix_ones
```

```python
plt.imshow(matrix_ones, cmap="Greys")
```

```python
matrix_ones=np.ones((8,8))
matrix_ones[:,::2]=0
matrix_ones
```

```python
plt.imshow(matrix_ones, cmap="Greys")
```

```python
matrix_ones=np.ones((8, 8))
matrix_ones[::2, ::2]=0

matrix_ones[1::2, 1::2]=0
```

```python
plt.imshow(matrix_ones, cmap="Greys")
```

```python
matrix_ones=np.ones(64)
matrix_ones[::2]=0
matrix_ones
```
