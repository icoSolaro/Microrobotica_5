```python
import numpy as np
import matplotlib.pyplot as plt
```

### Grafici a dispersione 
* con `plot()` non possiamo personalizzare i singoli punti
* con `scatter()` possiamo personalizzare i singoli punti

```python
N_PUNTI = 20

x=np.random.uniform(0., 1., N_PUNTI)
y=np.random.uniform(0., 1., N_PUNTI)
color=np.random.uniform(0., 1., N_PUNTI)
size=np.random.uniform(10., 1000., N_PUNTI)
alpha=np.random.uniform(0., 1., N_PUNTI)

fig, ax = plt.subplots(figsize=(4, 4))
ax.scatter(x, y, c=color, s=size, cmap="magma", alpha=alpha, edgecolor="k", marker="$$")#
plt.show()
```
