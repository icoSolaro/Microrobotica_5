```python
import numpy as np
import matplotlib.pyplot as plt

img_read=plt.imread("./EsempioImg.jpg")
img=img_read.copy()  # faccio deep copy per poter poi modificare immagine
plt.imshow(img)
```

```python
# indicizzazione booleana.
# data un immagine dove supera una certa soglia lo saturiamo
thr=100 # soglia
np.count_nonzero(img[:, : , 0]>thr) # canale rosso, quanti pixel ci sono con almeno 100 di soglia?
```

```python
red_thr=img[:, : , 0]>thr

img[red_thr]=np.array([255, 0, 0])
plt.imshow(img)
```
