### Es 1
Creare una matrice 4x5 di numeri casuali. sottrai da ogni colonna la sua media e dividi per la sua deviazione standard (np.std)



```python
matrix = np.random.uniform(0, 100, (4, 5))
matrix
```

```python
mean=np.mean(matrix, axis=0)
mean
```

```python
matrix_standard=(matrix-mean)/np.std(matrix, axis=0)
matrix_standard
```

### es 2
data una matrice 6x6 di interi casuali tra -10 e 10, sostituisci tutti i valori negativi con 0, estrai solo le righe che hanno una somma maggiore di 15

```python
matrix = np.random.randint(-10, 11, (6, 6))
matrix[(matrix<0)]=0
rows=matrix[(np.sum(matrix, axis=1))>15]
rows
```

### Es 3
dato un array di 100 temperature giornaliere casuali tra -5 e 35 °C: trova quanti giorni hanno temperatura tra 15 e 25 °C, sostituisci le temperature sotto 0 con 0 e sopra 30 con 30

```python
temperature=np.random.uniform(-5, 35, 100)
temp15_25=np.count_nonzero((temperature>15) & (temperature<25))
temp15_25
```

```python
temperature[temperature<0]=0
temperature[temperature>30]=30
temperature
```
