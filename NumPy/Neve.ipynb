```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_PUNTI=150
WHITES=['#DAE2F2', '#F2F2F2', '#E0EBFF']


x=np.random.uniform(0., 1., N_PUNTI)
y=np.random.uniform(0., 1., N_PUNTI)


s=np.clip(np.random.normal(100, 60, N_PUNTI), a_min=10, a_max=200)
#s pu conterere numeri negativi perch la funzione gaussiana, usiamo clip che assegna ai valori esterni ad un intervallo il valore dell'estremo
a=np.ones(N_PUNTI, dtype=np.float32)
# tutti i fiocchi pi piccoli di 120 hanno trasparenza 0.3, i rimanenti sono opachi(alpha=1)
a[s<120]=0.3
c=np.random.choice(WHITES, N_PUNTI)

fig, ax=plt.subplots(figsize=(10, 10))
ax.scatter(x, y, c=c, s=s, marker='$*$', edgecolors='k', linewidth=0.1, alpha=a)
ax.set_facecolor('#252850')
plt.show()
```
