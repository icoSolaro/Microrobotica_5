### Lancio di un dado
Simuliamo `n` lanci di un dado a 6 facce.

```python
n=10
dice=np.random.randint(1, 7, n)
dice
```

```python
# quanti 6 abbiamo ottenuto?
six=dice==6
six
#six.dtype
#six.shape
```

```python
np.count_nonzero(six)
```

```python
# Vinco se faccio 6 oppure 5
win=(dice==6) | (dice==5)
```

```python
np.count_nonzero(win)
```

### Esercizio
Alice e Bob lanciano ciascuno un dado per `n` volte, ogni volta vince chi fa il lancio più alto, Chi vince tra i due?

```python
n=10
a=np.random.randint(1, 7, n)
b=np.random.randint(1, 7, n)

print(f"Alice vince {np.count_nonzero(a>b)} volte")
print(f"Bob vince {np.count_nonzero(a<b)} volte")
print(f"Alice e Bob pareggiano {np.count_nonzero(a==b)} volte")
```

```python
# Ora Alice e Bob lanciano 2 dadi
n=10
a1=np.random.randint(1, 7, n)
b1=np.random.randint(1, 7, n)
a2=np.random.randint(1, 7, n)
b2=np.random.randint(1, 7, n)

totA=a1+a2
totB=b1+b2

print(f"Alice vince {np.count_nonzero(totA>totB)} volte")
print(f"Bob vince {np.count_nonzero(totA<totB)} volte")
print(f"Alice e Bob pareggiano {np.count_nonzero(totA==totB)} volte")
```
