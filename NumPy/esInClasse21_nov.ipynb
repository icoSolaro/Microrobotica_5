creare una matrice di dimensione `SIZE` che dentro abbia `N` uni posizionati a caso, gli atri elementi sono tutti 0

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
SIZE=(100, 151)
N=900

array1d=np.zeros(SIZE[0]*SIZE[1])

array1d[np.random.choice(np.arange(SIZE[0]*SIZE[1]), N, replace=False)]=1

matrix=array1d.reshape(SIZE)
matrix

#plt.imshow(matrix, cmap='gray')
plt.imshow(matrix, cmap=ListedColormap(["#252850", "white"]))

```
