```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generare un array 1d di valori equispaziati in un intervallo
np.linspace(0., 5., 101) # genera x(101) valori quispaziati tra 0 e 5 con 5 incluso, con parametro endpoint=False toglie estremo 
```

Numpy implementa tutte le principali funzioni matematiche. Per esempio la funzione $e^x$ è implementata da `np.exp(x)`.
La gaussiana semplice ha espressione $e^{-x}$

```python
def gaussiana_semplice(x):
    return np.exp(-x**2)

x=np.linspace(-10., 10., 1000)# ascissa    calcola brutalmente per 1000 punti i valori in y e li salva nell'array y
y=gaussiana_semplice(x)# ordinata         broadcast di numPy

plt.plot(x, y)
```

### I GRAFICI
fig è la tela(canva) su cui faccio i grafici
ax sono gli assi di un grafico posto sulla figure, se ho un solo grafico è un oggetto, se voglio più grafici sulla stessa fig ax è un array.
figsize è la dimensione della tela in pollici

```python
fig, ax = plt.subplots(figsize=(5,2))     #fig e ax sono nomi usati globalmente
fig.set_facecolor("#f2f3ae")  
ax.set_facecolor("#EDD382")  
ax.plot(x, y, color="#020122", linewidth=2., label=r"$f(x)=e^{-x^2}$")
ax.grid()
ax.legend(loc="upper left")   # loc per spostare la posizione della legenda
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Il primo grafico in Matplotlib")
plt.show()
```

```python
f=lambda x:np.sin(x)/x
g=lambda x:np.cos(x)
```

```python
x=np.linspace(-4*np.pi, 4*np.pi, 2000)
y1=f(x)
y2=g(x)
```

```python
fig, ax = plt.subplots(figsize=(10,5))     #fig e ax sono nomi usati globalmente
fig.set_facecolor("#f2f3ae")  
ax.set_facecolor("#EDD382")  
ax.plot(x, y1, color="#F4442E", linewidth=2., label=r"$f(x) = \frac{\sin(x)}{x}$")
ax.plot(x, y2, color="#7B2328", linewidth=2., label=r"$f(x)=cos(x)$")
ax.grid()
ax.legend(loc="upper right")   # loc per spostare la posizione della legenda
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("")
ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["top", "right"]].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
plt.show()
```

Dato un array qualsiasi di numeri verificare se è monotono crescente.

```python
(np.roll(monotono, -1) > monotono)[:-1] # per controllare se è monotono devo fare 4 controlli, il primo elemento non è confrontato con altro basta sia minore dei quello dopo
```

```python
# np.any() fa una or tra tutti gli elementi
np.all((np.roll(monotono, -1) > monotono)[:-1]) # all fa una and tra tutti. restituisce true se tutto true
```
