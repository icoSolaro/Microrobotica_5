```python
import random

class AutoregressioneDiMarkov:
    def __init__(self, txt, l_context=2):
        self.l_context=l_context
        file=open(txt, "r", encoding="utf-8")
        self.text=file.read()
        self.markov_dict={} #Contiente chiave n lettere con n=l_context e valore lista dei successivi caratteri. 
        self.preProcessing()

    def preProcessing(self):
        self.text=self.text.lower()
        self.text=self.text.replace('\n', ' ')
        for _ in range(3):
            self.text=self.text.replace('  ', ' ')

    def setDictionary(self):
        # i nostri token sono sempre composti da un singolo carattere carattere
        #un LLM tiene conto di 200 mila token, noi teniamo conto, di default,  di due token dove ogni token per noi è solo una lettera.
        #Il parametro che imposta il numero di token di cui teniamo conto è `markov_dict`
        
        for i, _ in enumerate(self.text[:-self.l_context]): #ciclando sul testo con indice carattere. Non deve arrivare in fondo perchè devo accedere alle posizioni successive all'attuale, quindi andrei oltre alla lunghezza massima
            ngram=self.text[i:i+self.l_context]#slicing per avere un ngramma che contiene dalla posizione attuale della stringa fino al carattere di l_context posizioni successive
            if ngram in self.markov_dict:
                self.markov_dict[ngram].append(self.text[i+self.l_context])# vado a salvarmi la lettera succesiva all'ennegramma che sto analizzando
            else:
                self.markov_dict[ngram]=[self.text[i+self.l_context]]

    def prediction(self, l_phrase=100, ngram_0=''):#l_phrase=600 contiene la lunghezza della frase.
        if ngram_0=='':
            ngram_0=self.text[:self.l_context]
        phrase=ngram_0# in phrase accumulo, creo tutta la frase
        
        if len(ngram_0)!=self.l_context:
            print("Errore!")
        else:
            for _ in range(l_phrase):
                if ngram_0 in self.markov_dict:
                    next_char=random.choice(self.markov_dict[ngram_0])
                    phrase=phrase+next_char
                    ngram_0=phrase[-self.l_context:]
                else:
                    print("Ngramma non presente")
                    break
        
        return phrase

    
autoregressioneDiMarkov=AutoregressioneDiMarkov("./data/sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt", 5)
autoregressioneDiMarkov.setDictionary()
phrase=autoregressioneDiMarkov.prediction()
print(phrase, 'il nu')
```
