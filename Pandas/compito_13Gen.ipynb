## compito griglia 7x7
Partendo dal dataset fer2013.csv creare un grafico con 7 x7 sistemi di assi in cui ogni riga rappresenta 7 facce scelte a caso. Le facce di ogni riga hanno la stessa emozione.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./fer2013.csv")
emotion_dict={0:"Angry", 1:"Disgust", 2:"Fear", 3:"Happy", 4:"Sad", 5:"Surprise", 6:"Neutral"}
```

```python
fig, axs=plt.subplots(ncols=7, nrows=7, figsize=(20, 20))
for i in emotion_dict:
    emotion=data[data["emotion"]==i]
    faces_array=emotion["pixels"].to_numpy()
    test_faces=np.random.choice(faces_array, 7, replace=False)
    for j, test_face in enumerate(test_faces):
        test_faces_array=np.array([int(p) for p in test_face.split(" ")]).astype(np.uint8)
        axs[i, j].imshow(test_faces_array.reshape((48, 48)), cmap="binary_r")
        axs[i, j].set_axis_off()
    axs[i, 3].set_title(emotion_dict[i], fontsize=20)

plt.show()
```
