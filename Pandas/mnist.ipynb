```python
fig, axs = plt.subplots(ncols=10, nrows=10, figsize=(30, 30))

for i in range(10):
    nums=data[data["class"]==i]
    pixel_columns_byNum=list(nums.columns)[:-1]
    images_byNum=nums[pixel_columns_byNum].to_numpy().astype(np.uint8)
    test_images=images_byNum[np.random.choice(len(images_byNum), 10, replace=False)]
    for j in range(10):
        axs[i,j].imshow(test_images[j].reshape((28, 28)), cmap="binary")
        axs[i,j].set_axis_off()
    
plt.show()
```
