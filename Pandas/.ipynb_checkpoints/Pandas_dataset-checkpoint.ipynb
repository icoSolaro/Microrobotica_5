```python
data=pd.read_csv("./penguins_size.csv")
```

```python
data.columns
```

```python
cols= ['species', 'culmen_length_mm', 'culmen_depth_mm','flipper_length_mm', 'body_mass_g']
```

```python
selected_data=data[cols].dropna()
selected_data.head()
```

```python
selected_data.info() #ho perso pochi valori anche perchè il sesso ha molti valori NaN
```

```python
# tutti i pinguini di specie Adelie che pesano meno di 4000g
selected_data[(selected_data["species"]=='Adelie') & (selected_data["body_mass_g"]<4000)]
```

```python
selected_data["body_mass_kg"]=selected_data["body_mass_g"]/1000
selected_data.head(5)
```

```python
dati_selezionati=selected_data.rename(columns={"species":"specie", "culmen_length_mm":"lunghezza_becco_mm", 
                                               "culmen_depth_mm":"profondita_becco_mm", "flipper_length_mm":"lunghezza_pinna_mm", 
                                               "body_mass_g": "peso_g", "body_mass_kg": "peso_kg"})
dati_selezionati
```

```python
dati_selezionati["peso_kg"].to_numpy() #trasforma in array numpy
```

```python
# Funzioni di aggregazione
selected_data["body_mass_kg"].max() # SELECT MAX(body_mass_kg) FROM selected_data
```

```python
selected_data["body_mass_kg"].mean()
```

```python
selected_data["species"].unique()
```

```python
# quale è il peso medio di ogni specie di pinguino? Ovvero come fare un group by
selected_data[["species", "body_mass_kg"]].groupby("species").mean()
```

```python
selected_data.groupby("species").mean() # raggruppa per specie e fa la media su ogni.
```

```python
#Ora vogliamo il peso medio dei pinguini aggregato per specie e per sesso.
data[["species", "sex", "body_mass_g"]].groupby(["species", "sex"]).mean()
```

```python
data[["species", "body_mass_g"]].value_counts()
```
