```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./fer2013.csv") #dataset di una gara di machinelearning che doveva riconoscere espressioni facciali. la prima colonna è l'emozione (0=arrabbiato) la seconda colonna sono immagini 48x48 
```

### `data` è un **DataFrame**

è un df di tre colonne, quella tutta a sinistra è un indice ed è gestito in maniera diversa dalle colonne

```python
# stampiamo le prime 5 righe del DataFrame data
data.head(5)  #le ultime con tail
```

```python
data.info()
```

```python
# come si accede a una colonna
data["emotion"]
```

```python
emotion_dict={0:"Angry", 1:"Disgust", 2:"Fear", 3:"Happy", 4:"Sad", 5:"Surprise", 6:"Neutral"}
```

```python
#selezionare tutti i record con emotion = 0 quindi arrabbiati
angry=data[data["emotion"]==0]
```

```python
angry
```

```python
angry_face_array=angry["pixels"].to_numpy()
#angry_face_array
```

```python
angry_face_array.shape
```

```python
angry_face_array.dtype
```

```python
test_face=angry_face_array[0]
```

```python
type(test_face)
```

Si tratta di immagini in scala di grigi 48x48

```python
# test_face.split(" ") # lista di stringhe
test_face_array=np.array([int(p) for p in test_face.split(" ")]).astype(np.uint8) #list comprension per trasformare in itero
test_face_array
```

```python
fig, ax=plt.subplots()
ax.imshow(test_face_array.reshape((48, 48)), cmap="binary_r")
ax.set_axis_off()
plt.show()
```

```python
fig, axs=plt.subplots(ncols=1, nrows=7, figsize=(2, 20))
for i in emotion_dict:
    emotion=data[data["emotion"]==i]
    faces_array=emotion["pixels"].to_numpy()
    test_face=faces_array[0]
    test_faces_array=np.array([int(p) for p in test_face.split(" ")]).astype(np.uint8)
    axs[i].imshow(test_faces_array.reshape((48, 48)), cmap="binary_r")
    axs[i].set_axis_off()
    axs[i].set_title(emotion_dict[i])

plt.show()
```
