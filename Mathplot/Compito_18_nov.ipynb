```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.sin(x)**2
g = lambda x: np.exp(-x**2)

x = np.linspace(-5, 5, 100000)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_facecolor('#F5F0E8')

ax.plot(x, f(x), color='#D55E00', linewidth=2, label=r'$f(x) = (\sin x)^2$')
ax.plot(x, g(x), color='#8FBC3B', linewidth=2, linestyle='--', label=r'$g(x) = e^{-x^2}$')

ax.grid()

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title(r'Grafico delle funzioni $f(x)$ e $g(x)$')

ax.legend(loc='upper right')

plt.show()
```
