```python
import numpy as np
import matplotlib.pyplot as plt
```

## Grafici: istrogrammi
Gli istogrammi **NON** sono grafici a barre!

```python
N_PERSONE = 1000
altezze=np.round(np.random.normal(170, 12, N_PERSONE), 1) #round(array, numero) arrotonda i valoria a una cirfra dopo la virgola: 1cm
altezze
```

```python
fig, ax=plt.subplots(figsize=(4, 3))
ax.hist(altezze, bins=20) #bins = in questo caso è il n° di intervalli dell'ascissa, quindi il n° di barre
ax.set_xlabel("altezza (cm)")
ax.set_ylabel("n° di persone")
plt.show
```

```python
N_LANCI=100000
dado1=np.random.randint(1, 7, N_LANCI)
dado2=np.random.randint(1, 7, N_LANCI)

punteggio=dado2+dado1    # ho 11 possibilità disponibili

fig, ax=plt.subplots(figsize=(4, 3))
ax.hist(punteggio, bins=11, alpha=0.5, color="red", edgecolor="k") #bins = in questo caso è il n° di intervalli dell'ascissa, quindi il n° di barre
#alpha=trasparenza
ax.hist(dado1, bins=6, alpha=0.5, color="blue", edgecolor="k") #bins = in questo caso è il n° di intervalli dell'ascissa, quindi il n° di barre
ax.set_xlabel("punteggio")
ax.set_ylabel("n° di volte che è uscito il punteggio")
ax.set_xticks(np.arange(2, 13))
plt.show
```
