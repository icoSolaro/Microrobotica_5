```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=plt.imread("./EsempioImg.jpg")
plt.imshow(img)
```

## figure con pi sistemi di assi


```python
fig, axs=plt.subplots(nrows=2, ncols=2, figsize=(12, 8))#axs  un array numpy con shape nrows, ncols. ogni elemento  un sistema di assi
axs[0,0].imshow(img)
axs[0,0].set_axis_off()


axs[0,1].imshow(img[:, :, 0], cmap="Reds")#img[:, :, 0]  una matrice, devo dirgli di associare i valri dei pixel a un colore. un immagine RGB SONO 3 MATRICI
axs[0,0].set_axis_off()


axs[1, 0].imshow(img[:, :, 1], cmap="Greens")#img[:, :, 0]  una matrice, devo dirgli di associare i valri dei pixel a un colore. un immagine RGB SONO 3 MATRICI
axs[0,0].set_axis_off()


axs[1, 1].imshow(img[:, :, 2], cmap="Blues")#img[:, :, 0]  una matrice, devo dirgli di associare i valri dei pixel a un colore. un immagine RGB SONO 3 MATRICI
axs[0,0].set_axis_off()




plt.show()
```
