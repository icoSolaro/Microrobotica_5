```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_LANCI=100000
N_DADI=2000
matrixDadi=np.random.randint(1, 7, (N_LANCI, N_DADI))#N_DADI righe
matrixDadi
```

```python
punteggio=np.sum(matrixDadi, axis=1)


fig, ax=plt.subplots(figsize=(4, 3))
ax.hist(punteggio, bins=60, alpha=1, color="red", edgecolor="k") 
ax.set_xlabel("punteggio")
ax.set_ylabel("n° di volte che è uscito il punteggio")

plt.show
```

```python
N_DADI1=10
N_DADI2=100
N_DADI3=10000


def lancio_Dadi(n_dadi):
    matrix=np.random.randint(1, 7, (N_LANCI, n_dadi))
    return np.sum(matrix, axis=1)

fig, ax=plt.subplots(figsize=(10, 11))
ax.hist(lancio_Dadi(N_DADI1), bins=60, alpha=0.4, color="red", edgecolor="k") 
ax.hist(lancio_Dadi(N_DADI2), bins=60, alpha=0.4, color="blue", edgecolor="k") 
ax.hist(lancio_Dadi(N_DADI3), bins=60, alpha=0.4, color="green", edgecolor="k") 

ax.set_xlabel("punteggio")
ax.set_ylabel("n° di volte che è uscito il punteggio")

plt.show
```

```python
def lancio_Dadi(n_dadi):
    matrix=np.random.randint(1, 7, (N_LANCI, n_dadi))
    return np.sum(matrix, axis=1)
```

```python
dadi=[2, 10, 100]
n_beans=[(6 * n) - n + 1 for n in dadi]
print(n_beans)
colors=["red", "blue", "green"]
fig, ax=plt.subplots(figsize=(10, 11))
for dado, beans, color in zip(dadi, n_beans, colors):
    ax.hist(lancio_Dadi(dado), bins=beans, alpha=0.4, color=color)

ax.set_xlabel("punteggio")
ax.set_ylabel("n° di volte che è uscito il punteggio")

plt.show
```
