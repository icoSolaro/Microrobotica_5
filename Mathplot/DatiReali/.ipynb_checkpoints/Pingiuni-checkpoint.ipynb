```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('C:/Micro/Mathplot/DatiReali/penguins_size.csv')
data # è un oggetto di tipo DataFrame
```

vogliamo grafici con questi dati

```python
data.dropna(inplace=True)    #elimina le righe dove almeno un dato è NaN. NON ANDREBBE FATTO COSI
data
```

```python
data_array=data[['culmen_length_mm', 'culmen_depth_mm',	'flipper_length_mm', 'body_mass_g']].to_numpy() #doppie [] perchè passo una lista di colonne. passo una lista di nomi di colonne. prima quadra per accedere a chiave, seconda per dire che è una lista
species=data['species'].to_numpy()
# to numpy trasforma un DF in un array np
```

```python
species
```

```python
specie2num={"Adelie":0, "Chinstrap":1, "Gentoo": 2}
species_num=np.zeros(len(species), dtype=np.uint8)
for s in specie2num:
    species_num[species==s]=specie2num[s]

species_num
```

```python
data_array.shape
```

```python
data_array.dtype
```

grafico a dispersione che fa vedere la dipendenza della grandezza del becco dal peso


```python
x = data_array[:, 2]
y = data_array[:, 3]
z = data_array[:, 1] #dimensione becco


fig, axs=plt.subplots(nrows=2, ncols=2, figsize=(8, 8))
axs[0,0].scatter(x, y, edgecolor='k', alpha=0.7, c=species_num)
axs[0,0].set_xlabel("flipper length (mm)")
axs[0,0].set_ylabel("body mass (g)")

axs[0,1].scatter(x, z, edgecolor='k', alpha=0.7, c=species_num)
axs[0,1].set_xlabel("flipper length (mm)")
axs[0,1].set_ylabel("culmen depth mm (mm)")

axs[1,0].hist(y, bins=50, color="red", edgecolor="k") 
axs[1,0].set_ylabel("body mass (g)")

axs[1,1].hist(x, bins=50, color="red", edgecolor="k") 
axs[1,1].set_ylabel("body mass (g)")

plt.show()
```
