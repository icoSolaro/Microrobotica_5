## ESERCIZIO 1
Apri l'immagine `image.jpg` come array di Numpy. Crea una funzione che utilizzi NumPy per creare una immagine compressa nel seguente modo:
- rimuovendo tutte le righe dispari (0 è considerato pari);
- rimuovendo tutte le colonne dispari (0 è considerato pari).

L'immagine ottenuta avrà un numero di righe pari alla metà di quelle iniziali, un numero di colonne pari alla metà di quelle iniziali. La funzione restituisce una deep copy dell'immagine compressa in tal modo.

Crea una figura con 4 sistemi di assi, dove visualizzi:
* l'immagine originale
* l'immagine compressa una volta
* l'immagine compressa due volte
* l'immagine compressa tre volte

**NON PUOI USARE CICLI, IF E LISTE**

```python
def comprimi(img):
    imgCmpressa=img[::2, ::2, :]
    return imgCmpressa

img=plt.imread('./image.jpg')

# print(img.shape)
# print(comprimi(img).shape)

fig, axs=plt.subplots(figsize=(15, 15), nrows=2, ncols=2)
axs[0,0].imshow(img)
axs[0,1].imshow(comprimi(img))
axs[1,0].imshow(comprimi(comprimi(img)))
axs[1,1].imshow(comprimi(comprimi(comprimi(img))))
```

## ESERCIZIO 2
* Dato `n` numero intero > 2, crea un array bidimensionale `n x n`, nel quale le righe pari hanno valori pari a 0, quelle dispari hanno un valore pseudo-casuale compreso tra 0. e 1.. 

* Infine tutti gli elementi aventi un valore maggiore di 0.5 devono essere posti a 0. 

* Calcola la media di ogni riga.

**NON PUOI USARE CICLI, IF E LISTE**

```python
N=6
matrix=np.zeros((N,N))
matrix[1::2,:]=np.random.random((N//2, N))
matrix[matrix>0.5]=0
np.mean(matrix, axis=1)
```

## ESERCIZIO 3
Simula 100 partite di pari e dispari tra Alice e Bob. Ogni partita è giocata con una mano, quindi i numeri mostrati da Alice e da Bob sono sempre compresi tra 1 e 5. **Simula le partite usando un solo array bidimensionale.**

Alice dichiara sempre pari e Bob sempre dispari.

1) Quante volte vince Alice?

2) Se nelle 100 partite Alice fa uscire sempre 3, quante volte vince?

**NON PUOI USARE CICLI, IF E LISTE**

```python
N_PARTITE=100
MIN=1
MAX=5
partite_random=np.random.randint(MIN, MAX+1, N_PARTITE)+np.random.randint(MIN, MAX+1, N_PARTITE)

print(f"Alice vince {np.count_nonzero((partite_random+1)%2)} volte")

partite_A3=3+np.random.randint(MIN, MAX+1, N_PARTITE)

print(f"Alice, facendo uscire sempre 3, vince {np.count_nonzero((partite_A3+1)%2)} volte")
```

```python
N_PARTITE=100
MIN=1
MAX=5
N_LANCI=2

partite = np.random.randint(1, 6, (N_PARTITE, N_LANCI))
somma = np.sum(partite, axis=1)
n_aliceVittorie = np.count_nonzero(somma%2==0)

print(n_aliceVittorie)

partite[:, 0]=3

somma = np.sum(partite, axis=1)
n_aliceVittorie = np.count_nonzero(somma%2==0)

print(n_aliceVittorie)
```

## ESERCIZIO 4
Rappresenta la seguente funzione $f(x)$ in un grafico che sia completo di tutti gli elementi necessari. Puoi usare una sola chiamata a `plot()`.

$ f(x) =   \left\{
\begin{array}{ll}
      e^x & x\le 0 \\
      \cos(2\pi x) & x \gt 0 \\
\end{array} 
\right.  $

```python
INTERVALLO=(-10, 10)

def f(x):
    y = x
    y=np.exp(x)
    y[x>0]=np.cos(2*np.pi*x[x>0])
    return y

x = np.linspace(INTERVALLO[0], INTERVALLO[1], 100000)
y=f(x)

fig, ax=plt.subplots(figsize=(12,6))
ax.scatter(x, y, )
fig.set_facecolor('#FFFFF0')
ax.set_title(r"$f(x)$")
ax.set_facecolor('#FFEEEE')
#ax.plot(x, y, c='r', linewidth=1.2)
ax.grid()
ax.set_xlabel(r"$x$")
ax.set_ylabel(r"$y$")
plt.show()
```
