```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier # classificatore multi-layer-Perceptron
```

```python
def scatter_plot(ax, x_name, y_name, categoria):
    for cat in data[categoria].unique():
        data_cat=data[data[categoria]==cat]
        ax.scatter(data_cat[x_name], data_cat[y_name], s=5, label=cat)
    ax.set_xlabel(x_name)
    ax.set_ylabel(y_name)
    #ax.legend()

def histogram_plot(ax, x_name, categoria):
    for cat in data[categoria].unique():
        data_cat=data[data[categoria]==cat]
        ax.hist(data_cat[x_name], bins=15 , label=cat, alpha=0.5)
    ax.set_xlabel(x_name)
    ax.set_ylabel("Frequenza assoluta")
    #ax.legend()

def analizzaDataset(data, categoria, nomeFile):
    validCols = [data[d].dtype!="object" for d in data]
    cols_numeriche=data.columns[validCols]
    n=len(cols_numeriche)    
    fig, axs = plt.subplots(figsize=(32, 32), nrows=n, ncols=n)
    for i, col_1 in enumerate(cols_numeriche):
        for j, col_2 in enumerate(cols_numeriche):
            if i!=j:
                scatter_plot(axs[i, j], col_1 , col_2, categoria)
            else:
                histogram_plot(axs[i, j], col_1, categoria)
    
    fig.legend(data[categoria].unique(), loc="center left", fontsize="xx-large")
    plt.savefig(nomeFile)
    plt.show()
    
```

```python
feature_cols = ['Pclass', 'Age', 'Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Fare', 'Sex_numeric']
output=['Survived']
```

```python
X=data[feature_cols].to_numpy()
y=data[output].to_numpy()
```

## Normalizzazione
significa trasformare la feature (colonna) in modo che abbia media 0 e deviazione standard 1

```python
#NORMALIZZAZIONE una features significa fare in modo che abbia media 0 e deviazione standard 1
pclass=X[:,0]
#pclass-np.mean(pclass) #avrebbe media 0
#vogliamo che abbia deviazione standard 1. vogliamo che tutte le colonne siano con media 0 e deviazione standard ugale tra loro. Cos abbiamo mappato tutte le colonne sullo stesso range
pclass_normalizzata=(pclass-np.mean(pclass))/np.std(pclass)
data["pclass_normalizzata"]=pclass_normalizzata
```

### verificare che l'istogramma di pclass  uguale a istogramma di pclass_normalizzata

```python
fis, axs = plt.subplots(ncols=2, nrows=1, figsize=(10, 5))

histogram_plot(axs[0], "Pclass", "Survived")
for survived in data["Survived"].unique():
    data_survived = data[data["Survived"] == survived]
    axs[1].hist(data_survived["pclass_normalizzata"], bins=15, label=survived, alpha=0.5)
axs[1].set_xlabel("pclass_normalizzata")
axs[1].set_ylabel("Frequenza assoluta")
axs[1].legend()

plt.show()
```

### StandardScaler
Si possono normalizzare tutte le features in un colpo solo

```python
features_scaler=StandardScaler()
X_norm=features_scaler.fit_transform(X)
```

```python
X_norm.dtype
```

```python
# features_scaler.transform([1,])
```

```python
X_train, X_test, y_train, y_test= train_test_split(X_norm, y, test_size=0.33, random_state=42)
```

```python
neural_network = MLPClassifier(hidden_layer_sizes=(4,3), max_iter=1000) # rete neurale non addestrata
```

```python
neural_network.fit(X_train, y_train.reshape(-1)) #y_train  un array unoD in colonne, vuole le righe
#addestrare una funzione vuol dire trovare il minimo della funzione, visto che la funzione non  derivabile usa dei passsaggi numero di passaggi=max_iter
```

```python
neural_network.score(X_train, y_train)
```

## Accuratezza
L'accuratezza  un esempio di **metrica** per misurare la bont di un algoritmo di machine learning: corrisponde alla percentuale di casi corretti.
Solitamente l'accuratezza di test  pi bassa, ma non sempre.
l'accuratezza 0.5  molto bassa, con 2 categorie equivale a tirare a caso

```python
neural_network.score(X_test, y_test)
```

#### Temperatura
il modella linguistico predice la prossima parola con una percentuale che equivale ad un energia, probabilit alta energia piccola. Un gas caldo ha tante molecole in tanti stati equivalenti probabili e bassi.
Normalmente un LLM prende un token di energia minima, la parola pi probabile, se si aumenta la temperatura significa che si pu andare a scegliere tra energie pi alte quindi la probabilit pi bassa. LLM diventa pi creativo

fisica di Maxwel e bolsman

$$p = A \cdot e^{-\frac{E}{k_B T}}$$
dove:

dove:
- $p$ = probabilit
- $A$ = costante di normalizzazione
- $E$ = energia
- $k_B$ = costante di Boltzmann $(1.38064910231.380649 \times 10^{-23}$
1.3806491023 J/K)
- $T$ = temperatura
