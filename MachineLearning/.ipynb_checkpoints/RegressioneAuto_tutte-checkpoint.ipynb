```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
data=pd.read_csv("./Used_fiat_500_in_Italy_dataset.csv")
data.info()
data.head()
data.describe()
```

```python
model_labelencoder=LabelEncoder()
data["model_numeric"]=model_labelencoder.fit_transform(data["model"])
#sex_labelencoder.transform(["male", "female"])
```

```python
trasmission_labelencoder=LabelEncoder()
data["trasmission_numeric"]=trasmission_labelencoder.fit_transform(data["transmission"])
```

```python
data.columns
```

```python
feature=['engine_power', 'age_in_days', 'km','previous_owners', 'lat', 'lon', 'model_numeric','trasmission_numeric']
output="price"

X=data[feature].to_numpy()
y=data[output].to_numpy()
```

```python
x_train, x_test, y_train, y_test = train_test_split (X, y, random_state=42, test_size=0.3)
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
#Coefficiente di determinazione sui dati di training
regressor.score(x_train, y_train)
```

```python
#Coefficiente di determinazione sui dati di test
regressor.score(x_test, y_test)
```

```python
feature=['engine_power', 'age_in_days', 'km','previous_owners', 'lat', 'lon', 'model_numeric','trasmission_numeric']
engine_power=70
age_in_days=365
km=20_000
previous_owners=1
lat=44.03
lon=7.5
model = "pop"
trasmission='manual'
model_numeric=model_labelencoder.transform([model])[0]
trasmission_numeric=trasmission_labelencoder.transform([trasmission])[0]
```

```python
fiat_500=np.array([[engine_power, age_in_days, km,previous_owners, lat, lon, model_numeric,trasmission_numeric]])
```

```python
price=regressor.predict(fiat_500)
price
```
