```python
def fattoriale(n):
    if n<1:
        return 0
    elif n==1:
        return 1
    else:
        return n*fattoriale(n-1)

fattoriale(4)
```

```python
def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else: 
        return fibonacci(n-1)+fibonacci(n-2)

fibonacci(5)
```
