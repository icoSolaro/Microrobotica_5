```python
def findChar(text, char):#le funzioni vanno dcumentate con il tripli apice
    '''
    La funzione ritorna True se char  prensente in text, False altrimenti
    '''
    return char in text

findChar('ciao', 'z')
    
```

```python
#Nome funzione come fosse fariabile      parola chiave  paramentri   quello che scriverei nel return
findChar                             =   lambda         text, char: char in text

findChar('ciao', 'x')
```

## In python tutto  un oggettto, anche le funzioni 

```python
somma = lambda a, b: a+b
prodotto=lambda a, b: a*b
operazioni={'+': somma, '*':prodotto}  #  sbagliato mettere le tonde qui perch sarebbe una chiamata

operazioni
```

```python
op=input(' inserisci + per la somma, * per il prodotto: ')
a=float(input('Inserisci il primo numero'))
b=float(input('Inserisci il secondo numero'))

if op in operazioni:
    print(operazioni[op](a, b))  # una funzione, le funzioni hanno senso se le chiamiamo, per chiamare le funzioni dobbiamo mettere tonde e paramentri
else: 
    print('Operazione non valida')
```
