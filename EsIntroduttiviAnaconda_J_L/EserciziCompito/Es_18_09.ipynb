* Create un dizionario che ha come chiavi tutti i caratteri presenti nel file e come valori delle liste. La lista contiene tutti i caratteri successori della lettera usata come chiave. Va bene che la lista contenga doppioni. Es casa => c:[a], a:[s, ' ']

### es 1 con frequenza assoluta

```python
fileName='./data/I_promessi_sposi.txt'
d1={}
d2={}
```

```python
file=open(fileName, 'r')
text=file.read()

# for char in text:
#     #print(char)
#     d1[char]=0

# for char in text:
#     d1[char]=d1[char]+1

for char in text:
    if char not in d1:
        d1[char]=1
    else:
        d1[char]=d1[char]+1

print(d1)
```

### es 2, con caratteri successori

```python
file=open(fileName, 'r')
text=file.read()

for char in text[1:]:
    d2[char]=[]

previous=text[0]
for char in text:
    d2[previous].append(char)
    previous=char

print(d2)
```

```python
import random
```

```python
num=int(input('Inserisci il numero di lettere che vuoi vengano generate: '))
char=input('Inserisci il carattere iniziale: ')

letters=[]

for _ in range(num):
    letters.append(char)
    char=random.choice(d2[char])

print(f'La parola formata: {"".join(letters)}')
```
